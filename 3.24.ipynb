```python
df_target_Weight= df_target.drop(["Waist","Pulse"], axis=1)
```

```python
df= pd.concat([df_data, df_target_Weight], axis =1)
```

```python
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)

lm_features = ["Chins", "Situps", "Jumps"]

for i, feature in enumerate(lm_features):
    row = int(i/4) 
    col = i%4

    sns.regplot(x=feature, y='Weight', data=df, ax=axs[row][col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split

x_data=df.loc[:, ['Chins', 'Situps', 'Jumps']]
y_data=df.loc[:, 'Weight']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print("():", np.round(lr.coef_, 1)) #np: numpy, coef_:     
print("():", np.round(lr.intercept_, 1))  #intercept_: () 
```

```python
y_preds = lr.predict(X_test)
print(X_train.shape, X_test.shape)
```

```python
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
print(" :", lr.intercept_) # y 

#  (coefficient) :       
print(":", np.round(lr.coef_,1))
```

```python
y_test_pred=lr.predict(X_test)

plt.figure(figsize=(10,5))
plt.scatter(X_test['Situps'], y_test, label='y_test')
plt.scatter(X_test['Situps'], y_test_pred, c='r', label='y_pred')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show() 
```
